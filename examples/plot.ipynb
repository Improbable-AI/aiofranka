```python
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
data = np.load("./sysid.npz")

qpos = data['qpos']
qvel = data['qvel']
qdes = data['qdes']
# ctrl = data['ctrl']
```

```python
qpos.shape, qvel.shape, qdes.shape
```

```python
fig, axes = plt.subplots(7, 1, figsize=(8, 13))

for i in range(7):
    axes[i].plot(qpos[:, i], label='qpos')
    axes[i].plot(qdes[:, i], label='qdes')
    # axes[i].plot(qvel[:, i], label='qvel')
    # axes[i].plot()
    # axes[i].plot(ctrl[:, i], label='ctrl')
    axes[i].set_ylabel(f'Joint {i+1}')

    # add vertical line with gray dotted line at every 100 timesteps
    for t in range(0, qpos.shape[0], 200):
        axes[i].axvline(t, color='gray', linestyle='--', linewidth=1.0)

    axes[i].legend()
```
